```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
# Load data

df = pd.read_csv("shootings.csv")
# df = pd.read_csv("shootings.csv", index_col = "date_", parse_dates = True)
```

```python
df=df.rename(columns = {'dist':'District number'})
df
```

```python
#Finding unique values in columns
a = df['offender_injured'].unique()
a
```

```python
#Drop unwanted columns

drop_column1 = ["the_geom" , "the_geom_webmercator" , "objectid" , "dc_key" , "lat" , "lng" , "point_x" , "point_y", "code","officer_involved", "offender_deceased", "location", "year"]
shootings = df.drop(drop_column1, axis = 1)

shootings.head()
```

```python
#Cleaning columns

shootings['Fatality'] = np.where(shootings['fatal']==1.0, 'Yes', 'No') #Make a new column
shootings['Indoor/Outdoor'] = np.where(shootings['inside']==1.0, 'Indoor', 'Outdoor') #Make a new column
shootings['Latino'] = np.where(shootings['latino']==1.0, 'Yes', 'No') #Make a new column
shootings = shootings.drop('fatal', 1)
shootings = shootings.drop('inside', 1)
shootings = shootings.drop('outside', 1)
shootings = shootings.drop('latino', 1)

shootings
```

```python
# convert to timestamp
shootings['date_'] = pd.to_datetime(shootings['date_'])
```

```python
#Creting year and month based on timestamp

shootings['Year'] = shootings['date_'].dt.year
shootings['Month'] = shootings['date_'].dt.month
shootings
```

```python
#MATPLOTLIB

#PIE-CHART
sorted_counts = shootings['District number'].value_counts() #finding the size of each variable in the column
plt.pie(sorted_counts, labels = sorted_counts.index, startangle = 0,
        counterclock = False, shadow=True, radius = 3, wedgeprops = {'width' : 1.5});
```

## Conclusion

#### The Pie-plot  shows the  distribution of the police district of shooting.
#### Interpretation:
(1) The Police districts with most shooting incidents is #25 followed by #22 and #24




```python
#Analysis for dates between 2018 - 2020

shootings_sorted = shootings.sort_values(by = ["date_"] , ascending = True)
#shootings_sorted.head()

#reset index
shootings_reset = shootings_sorted.reset_index(drop = True)
shootings_reset.head()
```

```python
shootings_trimmed = shootings_reset.loc[ (shootings_reset['Year'] >= 2018) & (shootings_reset['Year'] <= 2020 ) ]
shootings_trimmed
```

```python
#SEABORN plot for data between 2018 - 2020

color_map1 = {"B": "#1f77b4", "W": "#ff7f0e" , "M" : "#D62728"}


g = sns.lmplot(
    x="District number",
    y="age",
    hue="race",
    data=shootings_trimmed,
    height=12,
    aspect=1,
    palette=color_map1,
    scatter_kws=dict(edgecolor="none", alpha=0.2),
);

g.grid=True
g.set(xlim = (2,41))
g.set(xticks=range(0,41,1))

g.set(ylim = (10,120))
g.set(yticks=range(0,120,2))
```

## Conclusion

#### Regression scatter plot shows demographics (Black and White) as victims of shooting

#### Interpretation:

(1) District 24 & 25 hace majority white victims of shooting as compared to other districts.

(2) The average age of black victims is 28 across all district numbers

(3) The average age of white victims changes across the district numbers.

(4) Not all districts have recorded shootings

```python
#Group by Offender injured

shootings_counts = shootings.groupby(['District number', 'offender_injured', 'sex'],as_index=False).size()
shootings_counts.head()
```

```python
# initialize the chart with the data
chart = alt.Chart(shootings_counts)

# define what kind of marks to use
chart = chart.mark_circle(size=60)

# encode the visual channels
chart = chart.encode(
    x=alt.X("District number:Q", scale=alt.Scale(zero=False)),
    y=alt.Y("size:Q", scale=alt.Scale(zero=False)),
    color="offender_injured:N",
    tooltip=["offender_injured:N", "sex:N", "District number:Q", "size:Q"],
)

# make the chart interactive
chart.interactive()
```

## Conclusion

#### Altair facet plot shows place of shooting (either indoors or outdoors) as victims of shooting plotted District Number and quantity

#### Interpretation:

(1) Less number of offenders are injured.

(2) The plot shows distribution of offender being deceased (Y or N). For example, district 25 has a very number of not deceased shooters than deceased shooters.

```python
shootings_Fatality = shootings_trimmed.groupby(['District number', 'Fatality', 'age', 'Month'],as_index=False).size()
shootings_Fatality
```

```python
# Setup a brush selection
brush = alt.selection(type='interval')

# The top scatterplot
points = alt.Chart().mark_point().encode(
    x=alt.X('District number:Q', scale=alt.Scale(zero=False)),
    y=alt.Y('age:Q', scale=alt.Scale(zero=False)),
    color=alt.condition(brush, 'Fatality:N', alt.value('lightgray'))
).properties(
    selection=brush,
    width=800
)



#the bottom bar plot
bars = alt.Chart().mark_bar().encode(
    x='count(District number):Q',
    y='District number:N',
    color='Fatality:N',
).transform_filter(
    brush.ref() # the filter transform uses the selection
                # to filter the input data to this chart
).properties(
width=800
)

plot = alt.vconcat(points, bars, data=shootings_Fatality) # vertical stacking
plot
```

## Conclusion

#### Altair plot shows place of fatality of shooting (Yes or No) as victims of shooting plotted with District Number aganist age

#### Interpretation:

(1) Less victims have died in shootings.

(2) The filter-plot combinations helps find out the fatality rate across the different districts of Philadelphia simultaneously.
